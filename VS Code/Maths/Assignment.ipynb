Question 01

```python
def count():
    solution = []
    for e in range(0, 101):  
        for h in range(0, 101 - e): 
            g = 100 - e - h 
            cost = 5 * e + 0.75 * h + 0.25 * g
            if cost == 100:  
                solution.append((e, h, g))
    return solution

tot_count = count()

print("All possible solutions:")
for solution in tot_count:
    print("Number of Elephants:", solution[0])
    print("Number of Horses:", solution[1])
    print("Number of Goats:", solution[2])
    print()
```

```python
def fish_count():
    solutions = []
    for x in range(1, 200):
        remain = x
        split = []
        for i in range(3):  
            if (remain - 1) % 3 != 0:  
                break
            friend_share = (remain - 1) // 3
            split.append(friend_share)
            remain = (remain - 1) * 2 // 3 
        else:
            solutions.append((x, split))
    return solutions

tot_count = fish_count()

for answer in tot_count:
    initial_fish, split = answer
    print("Initial number of fish:", initial_fish)
    print("Fish split among three friends A, B, and C:", split)
    print()
```

```python
def is_safe(board, row, col, n):
    for i in range(col):
        if board[row][i] == 1:
            return False

    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False
        
    for i, j in zip(range(row, n, 1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False

    return True

def solve_nq_util(board, col, n, solutions):
    if col == n:
        copy_board = ["".join(["ðŸ†€" if col == 1 else "." for col in row]) for row in board]
        solutions.append(copy_board)
        return

    for i in range(n):
        if is_safe(board, i, col, n):
            board[i][col] = 1

            solve_nq_util(board, col + 1, n, solutions)

            board[i][col] = 0  

def solve_n_queens(n):
    board = [[0 for _ in range(n)] for _ in range(n)]
    solutions = []
    solve_nq_util(board, 0, n, solutions)
    
    if not solutions:
        print("Solution does not exist")
        return False

    for sol in solutions:
        for row in sol:
            print(row)
        print("\n")
    print(f"Total solutions: {len(solutions)}")
    return True

solve_n_queens(8)
```

Question 04

```python
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

def lcm(a, b):
    return a * b // gcd(a, b)

def lcm_of_three(a, b, c):
    return lcm(lcm(a, b), c)

a_interval = 3
b_interval = 4
c_interval = 10

lcm_all_clocks = lcm_of_three(a_interval, b_interval, c_interval)

print("You would hear all clocks ring at the same time again after", lcm_all_clocks, "hours.")
```

```python
def count_paths(m, n):
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(m + 1):
        dp[i][0] = 1
    for j in range(n + 1):
        dp[0][j] = 1

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            dp[i][j] = dp[i - 1][j] + dp[i][j - 1]

    return dp[m][n]

ways = count_paths(7, 4)
print("Number of ways:", ways)
```
