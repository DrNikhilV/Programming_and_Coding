```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gaussian(x, mean=0, std_dev=1):
    return (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)
```

```python
def monte_carlo_simulation(num_samples, mean=0, std_dev=1):
    samples = np.random.normal(mean, std_dev, num_samples)

    estimated_mean = np.mean(samples)
    return estimated_mean, samples
```

```python
num_samples = 10000  # Number of samples for the simulation
estimated_mean, samples = monte_carlo_simulation(num_samples)

print(f"Estimated Mean: {estimated_mean}")
```

```python
plt.figure(figsize=(10, 6))
plt.hist(samples, bins=50, density=True, alpha=0.6, color='b', label='Samples Histogram')

x = np.linspace(-4, 4, 1000)
plt.plot(x, gaussian(x), color='r', label='True Gaussian Distribution')
plt.axvline(estimated_mean, color='k', linestyle='dashed', linewidth=1, label='Estimated Mean')

plt.title('Monte Carlo Simulation for Gaussian Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```
